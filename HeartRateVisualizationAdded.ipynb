```python
#Import Dependencies
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size':14})
import pandas as pd
import numpy as np
```

```python
#Import heart information csv into dataframe
heart_added_df = pd.read_csv('Resources/HealthInfoAdded.csv')
#heart_added_df.head()

heart_df = pd.read_csv('Resources/HeartInfo.csv')
#heart_df.head()
heart_added_df.head()
```

```python
#Scatter Plot of Resting Heart Rate by Mass

#Marker Size
mrkr_s = np.array(heart_df['Longevity (Years)'])*10
mrkr_s2 = np.array(heart_added_df['Longevity (Years)'])*10
x1 = np.array(heart_df['Mass (grams)'])
y1 = np.array(heart_df['Resting Heart Rate (BPM)'])
x2 = np.array(heart_added_df['Mass (grams)'])
y2 = np.array(heart_added_df['Resting Heart Rate (BPM)'])

fig = plt.figure(figsize=(18,12))
ax1 = fig.add_subplot(111)
#Plot the data
#heart_df.plot.scatter(x='Mass (grams)', y='Resting Heart Rate (BPM)', s=mrkr_s, alpha=.75, linewidth=1.5, figsize=(18,12), 
#                      color='red', edgecolor='maroon')

ax1.scatter(x=x1, y=y1, s=mrkr_s, alpha=.75, linewidth=1.5, color='red', edgecolor='maroon')
ax1.scatter(x=x2, y=y2, s=mrkr_s2, alpha=.25, linewidth=1.5, color='blue', edgecolor='navy')

#Titles
plt.title('Resting Heart Rate by Mass', fontsize=50)
plt.xlabel('Mass (grams)', fontsize=30)
plt.ylabel('Resting Heart Rate (BPM)', fontsize=30)

#Legend
plt.legend(('Creature:'
            '\n'
            'Larger Dots'
            '\n'
            'Live Longer',), fontsize=15, loc='upper right')

#Grid & Axis Values
plt.grid(True)
plt.tick_params(labelright=True)
plt.xscale('log')
plt.xlim(1e0, 500000000)

#Annotation of Datapoints
xaxis= np.array(heart_df['Mass (grams)'])
yaxis= np.array(heart_df['Resting Heart Rate (BPM)'])
variable_label = np.array(heart_df['Creature'])

for i, txt in enumerate(variable_label):
    plt.annotate(txt, (xaxis[i]+10, yaxis[i]+10))
    
xaxis2= np.array(heart_added_df['Mass (grams)'])
yaxis2= np.array(heart_added_df['Resting Heart Rate (BPM)'])
variable_label = np.array(heart_added_df['Creature'])

for i, txt in enumerate(variable_label):
    plt.annotate(txt, (xaxis2[i]+10, yaxis2[i]+10))    


#Save Figure
plt.tight_layout()
#plt.savefig('Images/Resting_Heart_Rate_by_Heart_Mass_Added.png')
```
