```python
#Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Import heart information csv into dataframe
heart_df = pd.read_csv('Resources/HeartInfo.csv')
#heart_df.head()
```

```python
#Scatter Plot of Resting Heart Rate by Mass

#Plot the data
heart_df.plot.scatter(x='Mass (grams)', y='Resting Heart Rate (BPM)', figsize=(18,12), color='red')

#Titles
plt.title("Resting Heart Rate by Heart Mass")
plt.xlabel('Heart Mass (grams)')
plt.ylabel('Resting Heart Rate (BPM)')

#Grid & Axis Values
plt.grid(True)
plt.tick_params(labelright=True)
plt.xscale('log')

```
